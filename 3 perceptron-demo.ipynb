```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_classification
from sklearn.linear_model import Perceptron

```

# Generate synthetic 3D data


```python
X, y = make_classification(n_samples=100, n_features=3, n_classes=2, n_clusters_per_class=1, n_redundant=0, random_state=42)
```

```python
# Initialize and train the perceptron
perceptron = Perceptron(max_iter=100, eta0=0.1, random_state = 42)
```

```python
perceptron.fit(X, y)
```

# Plot the 3D scatter plot


```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis', marker='o', edgecolors='k')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')

# Create a meshgrid to plot decision regions
xx, yy = np.meshgrid(np.linspace(X[:, 0].min(), X[:, 0].max(), 100),
                     np.linspace(X[:, 1].min(), X[:, 1].max(), 100))
zz = -(perceptron.coef_[0][0] * xx + perceptron.coef_[0][1] * yy + perceptron.intercept_) / perceptron.coef_[0][2]

# Plot decision regions
ax.plot_surface(xx, yy, zz, alpha=0.3, cmap='viridis')
plt.title('Perceptron Decision Boundary in 3D')
plt.show()
```
